```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("01_student_marks.csv")
df
```

```python
df["Total"] = df[['Maths','Science','English','History','Computer']].sum(axis=1)
df["Percentage"]=df['Total']/5
```

```python
df
```

```python
plt.bar("Total")
```

```python
def Grade(Percentage):
    if Percentage>=90:
        return 'A'
    elif Percentage>=75:
        return 'B'
    elif Percentage>=60:
        return 'C'
    elif Percentage>=45:
        return 'D'
    else:
        return 'Fail'
```

```python
df['Grade']= np.vectorize(Grade)(df["Percentage"])
```

```python
df
```

```python
Topper = df[df["Total"]==df["Total"].max()]
print("Topper : \n", Topper[['Roll No','Name','Total','Percentage']])
```

```python
Average=df[["Maths","Science","English","History","Computer"]].mean()
print("Average of subject is\n",Average)
```

```python
df['Grade'].value_counts()
```

```python
failed_students = df[df['Grade'] == 'F']

if failed_students.empty:
    print("No one failed")
else:
    print(failed_students)
```
