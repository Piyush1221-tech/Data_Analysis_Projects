```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("02_Cricket_data.csv")
df
```

```python
Most_winner = df["winner"].value_counts()
most_winning_team = Most_winner.idxmax()  # Team with max wins
most_win_count = Most_winner.max()        # Number of wins

print(f"\n Most Winning Team: {most_winning_team} with {most_win_count} wins")
```

```python
a = df["home_team"].value_counts()
b = df["away_team"].value_counts()
h = (a+b)
most_played_team = h.idxmax()
most_played_count= h.max()
print(f"\n Most match played by a team : {most_played_team} with {most_played_count} times. ")
```

```python
Most_player_of_the_Match = df["pom"].value_counts()
most_times_player =Most_player_of_the_Match.idxmax() 
most_time_no_of_pom = Most_player_of_the_Match.max()        

print(f"\n Most time player of the match is : {most_times_player} pom is {most_time_no_of_pom} times")
```

```python
Most_hosted_stadium = df["venue_name"].value_counts()
stadium_name =Most_hosted_stadium.idxmax() 
no_of_time_hosted = Most_hosted_stadium.max()        

print(f"\n Most time hosted stadium is : {stadium_name} its hosted {no_of_time_hosted} times")
```

```python
most_toss_win = df["toss_won"].value_counts()
most_toss_win
```

```python
df["match_and_toss_winner"]=df["toss_won"]==df["winner"]
result=df["match_and_toss_winner"].value_counts()

match_toss_win = result.get(True)
toss_but_not_match = result.get(False)

print(f"toss and match both win : {match_toss_win} times")
print(f"toss win but not match  : {toss_but_not_match} times")
```
