```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("03_netflix_titles.csv")
df
```

```python
print(df.isnull().sum())
```

```python


df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not available')
df.dropna(subset=['country'], inplace=True)  # dropna on df directly is okay



```

```python
# Content type count
# Find how many entries are Movies and how many are TV Shows.
df["type"].value_counts()
```

```python
# Top countries with the most Netflix content
# Find top 10 countries by the number of titles produced.

df["country"].value_counts().head(10)
```

```python
# Top countries with the most Netflix content
# Find top 10 countries by the number of titles produced.

top_countries = df['country'].value_counts().head(10)
print(top_countries)
```

```python
# Top directors
# Identify the 10 directors with the most Netflix titles.
top_director=df["director"].value_counts().head(10)
top_director
```

```python
# Most frequent genres
# Find which categories (from listed_in) are most common.
most_genre=df["listed_in"].value_counts().head(10)
most_genre
```

```python
# 1. Split the categories by comma, create a list of all categories
all_categories = df['listed_in'].str.split(', ').explode()

# 2. Count the occurrences of each category
category_counts = all_categories.value_counts()

# 3. Show top 10 most common categories
print(category_counts)
```

```python
# Content released per year
#Find the number of titles released each year (from release_year).
year_wise = df["release_year"].value_counts().sort_index()
year_wise
```

```python
# Longest and shortest titles
# Identify the titles with the longest and shortest names.
df["title_length"]=df["title"].apply(len)
longest = df.loc[df["title_length"].idxmax()]["title"]
shortest = df.loc[df["title_length"].idxmin()]["title"]
```

```python
# Filtering & Grouping Filter for Indian content
# Find all titles produced in India.
indian_titles =df[df["country"].str.contains("India",na=False)]
print(indian_titles[['title', 'type', 'release_year', 'director']])
```

```python
#Filter by year
#Find all shows added in a specific year (e.g., 2020).
df[df["release_year"]==2020]
```

```python
#Find all horror movies
#Filter titles that belong to the “Horror” genre.

horror_movie =df[(df["type"]=="Movie")& (df["listed_in"].str.contains("Horror",na=False))]
print(horror_movie[["title","country","release_year","listed_in"]])


```

```python
# Find all shows/movies by a specific director
# Example: All titles directed by “Christopher Nolan”.
sepecific_director =df[df["director"].str.contains("Christopher Nolan",na=False)]
print(sepecific_director[["title","type","country","release_year"]])
 # OR
specific_director =df[df["director"].str.contains("Christopher Nolan", na=False)]
print(specific_director[["title", "type", "country","release_year"]])
```

```python
# Group by country
# Group data by country and count the number of titles.
country_title_counts = df.groupby('country').size().sort_values(ascending=False)
print(country_title_counts)
```
